We will use the titanic dataset from Kaggle (https://www.kaggle.com/). This is a well-known dataset and we will use it for classification- if the passenger survived or passed away.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
training_DF = pd.read_csv('titanic_dataset_GBC.csv')
training_DF.head()  # return the first 5 rows
```

```python
training_DF.describe()
```

```python
print(len(training_DF))
```

```python
np.isnan(training_DF['Age'])  # to test if elements are NaN(not a number) or not.
```

```python
sum(np.isnan(training_DF['Age'])) # total count of True results (NAN elements)
```

```python
# training_DF['Age'].fillna(training_DF['Age'].mean(), assign=True) # replaces the NAN values witht the total average 
                                                                   # and it inplaces in the existing DataFrame

training_DF = training_DF.assign(Age=training_DF['Age'].fillna(training_DF['Age'].mean()))
```

```python
sum(np.isnan(training_DF['Age']))
```

```python
training_DF.drop('Cabin',axis=1,inplace=True) # eliminates row/column, axis=0 for row and 1 for column
```

```python
training_DF.dropna(inplace=True) # eliminates rows with missing values
```

```python
training_DF.info()
```

```python
ax = pd.get_dummies(training_DF['Sex'])  # Converts categorical variables into dummy/indicator variables
embark = pd.get_dummies(training_DF['Embarked'])
```

```python
training_DF.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True) 
```

```python
training_DF = pd.concat([training_DF,ax,embark],axis=1) # concatenates new entries
```

```python
training_DF.head()
```

```python
training_DF.drop(['female','C'],axis=1,inplace=True)  # simplification, eliminating redundant parameters
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(training_DF.drop('Survived',axis=1), 
                                                    training_DF['Survived'], test_size=0.30, 
                                                    random_state=101)
# Setting random_state a fixed value will guarantee that the same sequence of random numbers is generated each time you run the code.
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#logmodel = LogisticRegression(random_state=0, max_iter=100).fit(X_train, y_train)
logmodel = LogisticRegression(random_state=0, max_iter=10000).fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print('Accuracy=',(149+71)/(149+14+33+71))
```

```python
from sklearn.metrics import classification_report

# 生成分類報告
report = classification_report(y_test, predictions)
print(report)
```

```python
from sklearn.linear_model import SGDClassifier # Stochastic Gradient Descent, a linear classifier (SVM, logistic regression, a.o.) optimized by the SGD
SGD_model = SGDClassifier(loss="hinge", penalty="l2", max_iter=100)
```

```python
SGD_model.fit(X_train, y_train)
#SGDClassifier(max_iter=5)
```

```python
SGD_predictions = SGD_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,SGD_predictions))
```

```python
print('Accuracy=',(157+29)/(157+29+75+1))
```

```python
from sklearn.metrics import classification_report

# 生成分類報告
report = classification_report(y_test, SGD_predictions)
print(report)
```
