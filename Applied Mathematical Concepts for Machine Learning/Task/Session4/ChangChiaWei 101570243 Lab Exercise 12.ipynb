### The Gaussian window is defined by $ w(n)=\exp^{\frac{-1}{2}{\frac{n}{\sigma}}^2}$ where $n$ is the number of pixels in the output window

```python
import numpy as np
from scipy.fft import fft2, ifft2, ifftshift
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler


def gaussian_window(size, std):
    x = np.linspace(-size // 2, size // 2, size)  
    gauss = np.exp(-0.5 * (x / std)**2)  
    return gauss / gauss.sum()  


rows, cols = new_im.shape


kernel = np.outer(gaussian_window(rows, 5), gaussian_window(cols, 5))  


freq = fft2(new_im)  
kern = ifftshift(kernel)  
freq_kernel = fft2(kern)  
product = freq * freq_kernel  
im_out = ifft2(product).real  

scaler = MinMaxScaler(feature_range=(0, 255))
im_out = scaler.fit_transform(im_out)

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(new_im, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(im_out, cmap='gray')
plt.title('Blurred Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

# Question 1: Is this a high-pass or a low-pass Gaussian filter? By modifying slightly the code above, create a 3-pixel Gaussian kernel and output the resulting image. What do you observe when you narrow the size of the Gaussian wiindow?



```python
# type your answer here (that includes code)
import numpy as np
from scipy.fft import fft2, ifft2, ifftshift
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Create a Gaussian window function
def gaussian_window(size, std):
    x = np.linspace(-size // 2, size // 2, size)
    gauss = np.exp(-0.5 * (x / std) ** 2)
    return gauss / gauss.sum()

# Load and preprocess the image
rows, cols = new_im.shape

# Create a 3-pixel Gaussian kernel (narrow window)
kernel = np.outer(gaussian_window(rows, 3), gaussian_window(cols, 3))

# Fourier Transform processing
freq = fft2(new_im)  # Image Fourier Transform
kern = ifftshift(kernel)  # Shift kernel center to top-left
freq_kernel = fft2(kern)  # Kernel Fourier Transform
product = freq * freq_kernel  # Element-wise multiplication in frequency domain
im_out = ifft2(product).real  # Inverse Fourier Transform, take real part

# Scale the resulting image to [0, 255]
scaler = MinMaxScaler(feature_range=(0, 255))
im_out = scaler.fit_transform(im_out)

# Display the images
plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(new_im, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(im_out, cmap='gray')
plt.title('Result with 3-Pixel Gaussian'), plt.xticks([]), plt.yticks([])

plt.show()
```

### Question 2: What would a high-pass Gaussian filter be useful for in the context of image processing?  

```python
import numpy as np
from scipy.fft import fft2, ifft2, ifftshift
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Create a Gaussian window function
def gaussian_window(size, std):
    x = np.linspace(-size // 2, size // 2, size)
    gauss = np.exp(-0.5 * (x / std) ** 2)
    return gauss / gauss.sum()

# Load and preprocess the image (assuming new_im is already loaded as grayscale)
rows, cols = new_im.shape

# Create a 2D Gaussian kernel (low-pass filter)
kernel = np.outer(gaussian_window(rows, 5), gaussian_window(cols, 5))

# Fourier Transform of the image
freq = fft2(new_im)

# Create a low-pass filter in the frequency domain
kern = ifftshift(kernel)  # Shift kernel center to top-left
freq_kernel = fft2(kern)  # Fourier Transform of kernel

# Perform low-pass filtering
low_pass = ifft2(freq * freq_kernel).real

# Generate high-pass filtered image by subtracting the low-pass filtered image
high_pass = new_im - low_pass

# Scale the resulting image to [0, 255]
scaler = MinMaxScaler(feature_range=(0, 255))
high_pass_scaled = scaler.fit_transform(high_pass)

# Display results
plt.figure(figsize=(12, 6))

plt.subplot(131), plt.imshow(new_im, cmap='gray')
plt.title("Original Image"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(low_pass, cmap='gray')
plt.title("Low-Pass Filtered Image"), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(high_pass_scaled, cmap='gray')
plt.title("High-Pass Gaussian Filtered Image"), plt.xticks([]), plt.yticks([])

plt.show()
```
