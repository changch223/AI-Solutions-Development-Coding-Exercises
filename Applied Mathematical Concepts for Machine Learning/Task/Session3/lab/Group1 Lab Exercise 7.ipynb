We will use the titanic dataset from Kaggle (https://www.kaggle.com/). This is a well-known dataset and we will use it for classification- if the passenger survived or passed away.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
training_DF = pd.read_csv('titanic_dataset_GBC.csv')
```

```python
training_DF
```

```python
training_DF['Age'].fillna(training_DF['Age'].mean(), inplace=True)
```

```python
training_DF.drop('Cabin',axis=1,inplace=True)
```

```python
training_DF.dropna(inplace=True)
```

```python
sex = pd.get_dummies(training_DF['Sex'])
embark = pd.get_dummies(training_DF['Embarked'])
```

```python
training_DF.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
training_DF = pd.concat([training_DF,sex,embark],axis=1)
```

```python
training_DF.drop(['female','C'],axis=1,inplace=True)
```

```python
training_DF
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(training_DF.drop('Survived',axis=1), 
                                                    training_DF['Survived'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# let's explore SGDClassifier parameters

'''
SGDClassifier(
    loss='hinge',
    penalty='l2',
    alpha=0.0001,
    l1_ratio=0.15,
    fit_intercept=True,
    max_iter=1000,
    tol=0.001,
    shuffle=True,
    verbose=0,
    epsilon=0.1,
    n_jobs=None,
    random_state=None,
    learning_rate='optimal',
    eta0=0.0,
    power_t=0.5,
    early_stopping=False,
    validation_fraction=0.1,
    n_iter_no_change=5,
    class_weight=None,
    warm_start=False,
    average=False,
)
'''

# Let's form SGD models with variation in paameters loss and alpha
    # loss: 'hinge', 'log', and 'modified_huber'
    # alpha: 0.0001 and 0.001
    # explain and dicuss your findings
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

loss_values = ['hinge', 'log_loss', 'modified_huber']
alpha_values = [0.0001, 0.001]

results = []

for loss_fn in loss_values:
    for alpha_val in alpha_values:
        
        sgd_clf = SGDClassifier(
            loss=loss_fn,
            alpha=alpha_val,
            random_state=42,
            max_iter=1000,
            tol=1e-3
        )
        sgd_clf.fit(X_train_scaled, y_train)
        
      
        y_pred = sgd_clf.predict(X_test_scaled)
        acc = accuracy_score(y_test, y_pred)
        
    
        results.append({
            'loss': loss_fn,
            'alpha': alpha_val,
            'accuracy': acc
        })

for r in results:
    print(r)
```

```python
# log_loss + alpha=0.001 (0.6629) leads to higher accuracy.
# for this dataset, low regularization (alpha=0.001) combined with the Logistic Regression loss is more effective in balancing "model fit" and "avoiding overfitting."
```
