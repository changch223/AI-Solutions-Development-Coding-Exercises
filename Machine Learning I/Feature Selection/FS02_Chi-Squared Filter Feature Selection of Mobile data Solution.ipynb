### Import the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

### Read the mobile data

```python
#data = pd.read_csv("data/mobile_price_train.csv")
data = df
```

### Split the dataset into X and y

```python
X = data.iloc[:,0:20]
y = data.iloc[:,-1] 
```

### Sanity check

```python
X.shape, y.shape
```

### How many features

```python
X.shape[1]
```

### Apply SelectKBest class to extract top 10 best features

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
bestfeatures
```

### Train the model

```python
topfeatures = bestfeatures.fit(X,y)
```

```python
topfeatures.scores_
```

### Create dataframes from Scores and Features

```python
dfscores = pd.DataFrame(topfeatures.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

### Concatenate two dataframes together

```python
featureScores = pd.concat([dfcolumns,dfscores], axis=1)
featureScores
```

### Add column names as Specs and Scores for the above dataframe

```python
featureScores.columns = ['Specs', 'Score']
```

```python
featureScores
```

### Which are the best features?

```python
print(featureScores.nlargest(10,'Score')) 
```
