# Gaussian Mixture Models Vs K-Means
## This notebook outlines the application of Gaussian Mixture Models versus K-Means

# Dataset :

https://raw.githubusercontent.com/subashgandyer/datasets/main/gmm.csv

The dataset used for this problem is Weights and Heights of 500 people without any labels.

2 features
- Weight
- Height

No labels

### Import the necessary libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Load the dataset

```python
url = "https://raw.githubusercontent.com/subashgandyer/datasets/main/gmm.csv"
```

```python
data = pd.read_csv(url)
data
```

### Plot the data

```python
plt.figure(figsize=(10,10))
plt.scatter(data["Weight"],data["Height"])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Data Distribution')
plt.show()
```

# K-Means Model
- Import the K-Means Estimator
- Create the K-Means Estimator object with number of clusters specified
- Fit the model
- Predict on data
- Plot the data points with the predicted labels

```python
#{Code可能是錯的}
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)


pred = kmeans.predict(data)
frame = pd.DataFrame(data)
frame['cluster'] = pred
frame.columns = ['Weight', 'Height', 'cluster']


plt.figure(figsize=(10,10))
color=['blue','green','cyan', 'black']
for k in range(0,4):
    data = frame[frame["cluster"]==k]
    plt.scatter(data["Weight"],data["Height"],c=color[k])
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 生成示例數據（如果沒有 data，請用這個測試）
# data = np.random.rand(300, 2) * 100

# 訓練 K-Means 模型
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)  # 增加 `n_init` 以提高穩定性
kmeans.fit(data)

# 預測群集
pred = kmeans.predict(data)

# 建立 DataFrame，確保欄位名稱正確
frame = pd.DataFrame(data, columns=['Weight', 'Height'])
frame['cluster'] = pred  # 新增聚類結果

# 設定顏色
plt.figure(figsize=(10,10))
colors = ['blue', 'green', 'cyan', 'black']

# 遍歷 4 個群集，繪製散點圖
for k in range(4):
    cluster_data = frame[frame["cluster"] == k]  # 使用不同變數名稱
    plt.scatter(cluster_data["Weight"], cluster_data["Height"], c=colors[k], label=f'Cluster {k}')

plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

# Gaussian Mixture Models
- Import the GaussianMixture Estimator
- Create the GaussianMixture Estimator object with number of clusters specified
- Fit the model
- Predict on data
- Plot the data points with the predicted labels

### Load the dataset from the specified url

```python
data = pd.read_csv(url)
data
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4)
gmm.fit(data)

#predictions from gmm
labels = gmm.predict(data)
frame = pd.DataFrame(data)
frame['cluster'] = labels
frame.columns = ['Weight', 'Height', 'cluster']

plt.figure(figsize=(10,10))
color=['blue','green','cyan', 'black']
for k in range(0,4):
    data = frame[frame["cluster"]==k]
    plt.scatter(data["Weight"],data["Height"],c=color[k])
plt.show()
```
